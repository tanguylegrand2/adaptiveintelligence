## The Environment

You can find the environment in the file Chess_env, which contains the class Chess_env. To define an object, you need to provide the board size considered as input. In our example, size_board=4. 
Chess_env is composed by the following methods:

1. Initialise_game. The method initialises an episode by placing the three pieces considered (Agent's king and queen, enemy's king) in the chess board. The outputs of the method are described below in order.

     S $\;$ A matrix representing the board locations filled with 4 numbers: 0, no piece in that position; 1, location of the 
     agent's king; 2 location of the queen; 3 location of the enemy king.
     
     X $\;$ The features, that is the input to the neural network. See the assignment for more information regarding the            definition of the features adopted. To personalise this, go into the Features method of the class Chess_env() and change        accordingly.
     
     allowed_a $\;$ The allowed actions that the agent can make. The agent is moving a king, with a total number of 8                possible actions, and a queen, with a total number of $(board_{size}-1)\times 8$ actions. The total number of possible actions correspond      to the sum of the two, but not all actions are allowed in a given position (movements to locations outside the borders or      against chess rules). Thus, the variable allowed_a is a vector that is one (zero) for an action that the agent can (can't)      make. Be careful, apply the policy considered on the actions that are allowed only.
     

2. OneStep. The method performs a one step update of the system. Given as input the action selected by the agent, it updates the chess board by performing that action and the response of the enemy king (which is a random allowed action in the settings considered). The first three outputs are the same as for the Initialise_game method, but the variables are computed for the position reached after the update of the system. The fourth and fifth outputs are:

     R $\;$ The reward. To change this, look at the OneStep method of the class where the rewards are set.
     
     Done $\;$ A variable that is 1 if the episode has ended (checkmate or draw).
     
     
3. Features. Given the chessboard position, the method computes the features.

This information and a quick analysis of the class should be all you need to get going. The other functions that the class exploits are uncommented and constitute an example on how not to write a python code. You can take a look at them if you want, but it is not necessary.

```python
# PERFORM N_episodes=1000 EPISODES MAKING RANDOM ACTIONS AND COMPUTE THE AVERAGE REWARD AND NUMBER OF MOVES 

S,X,allowed_a=env.Initialise_game()
N_episodes=1000

# VARIABLES WHERE TO SAVE THE FINAL REWARD IN AN EPISODE AND THE NUMBER OF MOVES 
R_save_random = np.zeros([N_episodes, 1])
N_moves_save_random = np.zeros([N_episodes, 1])

for n in range(N_episodes):
    
    S,X,allowed_a=env.Initialise_game()     # INITIALISE GAME
    Done=0                                  # SET Done=0 AT THE BEGINNING
    i=1                                     # COUNTER FOR THE NUMBER OF ACTIONS (MOVES) IN AN EPISODE
    
    # UNTIL THE EPISODE IS NOT OVER...(Done=0)
    while Done==0:
        
        # SAME AS THE CELL BEFORE, BUT SAVING THE RESULTS WHEN THE EPISODE TERMINATES 
        
        a,_=np.where(allowed_a==1)
        a_agent=np.random.permutation(a)[0]

        S,X,allowed_a,R,Done=env.OneStep(a_agent)
        
        
        if Done:
            
            R_save_random[n]=np.copy(R)
            N_moves_save_random[n]=np.copy(i)

            break

        i=i+1                               # UPDATE THE COUNTER



# AS YOU SEE, THE PERFORMANCE OF A RANDOM AGENT ARE NOT GREAT, SINCE THE MAJORITY OF THE POSITIONS END WITH A DRAW 
# (THE ENEMY KING IS NOT IN CHECK AND CAN'T MOVE)

print('Random_Agent, Average reward:',np.mean(R_save_random),'Number of steps: ',np.mean(N_moves_save_random))

```

```python
# INITIALISE THE PARAMETERS OF YOUR NEURAL NETWORK 
# PLEASE CONSIDER TO USE A MASK OF ONE FOR THE ACTION MADE AND ZERO OTHERWISE IF YOU ARE NOT USING VANILLA GRADIENT DESCENT...
# WE SUGGEST A NETWORK WITH ONE HIDDEN LAYER WITH SIZE 200. 


S,X,allowed_a=env.Initialise_game()
N_a=np.shape(allowed_a)[0]   # TOTAL NUMBER OF POSSIBLE ACTIONS

N_in=np.shape(X)[0]    ## INPUT SIZE
N_h=200                ## NUMBER OF HIDDEN NODES
N_out=N_a              ## output size

## INITALISE YOUR NEURAL NETWORK...
W1 = np.random.uniform(0,1,(N_h, N_in))
W2 = np.random.uniform(0,1,(N_out, N_h))

# Initialize the biases

bias_W1 = np.zeros((N_h,))
bias_W2 = np.zeros((N_out,))


# HYPERPARAMETERS SUGGESTED (FOR A GRID SIZE OF 4)

epsilon_0 = 0.2     # STARTING VALUE OF EPSILON FOR THE EPSILON-GREEDY POLICY
beta = 0.00005      # THE PARAMETER SETS HOW QUICKLY THE VALUE OF EPSILON IS DECAYING (SEE epsilon_f BELOW)
gamma = 0.85        # THE DISCOUNT FACTOR
eta = 0.0035        # THE LEARNING RATE

N_episodes = 50000 # THE NUMBER OF GAMES TO BE PLAYED 

# SAVING VARIABLES
R_save = np.zeros([N_episodes, 1])
N_moves_save = np.zeros([N_episodes, 1])
```

```python
# dÃ©finition of some function

#definition of ReLu function
def relu(x):
    rep=np.empty(0)
    for u in x:
        rep=np.append(rep,max(0.0,u))
    return rep


#function for the exponential moving average
def EMA(L,days):
    k=2/(days+1)
    t0=L[0]
    EMA=np.empty(0)
    EMA=np.append(EMA,t0)
    for i in range(1,len(L)):
        m=L[i]*k+EMA[i-1]*(1-k)
        EMA=np.append(EMA,m)
    return EMA

def MA(L,days):
    MA=np.empty(0)
    t0=L[0]
    MA=np.append(MA,t0)
    for i in range(1,len(L)):
        m=L[i]*k+EMA[i-1]*(1-k)
        EMA=np.append(EMA,m)
```

```python
#training with SARSA algorithm

#we do this for every game
for n in range(N_episodes):

    epsilon_f = epsilon_0 / (1 + beta * n)   ## DECAYING EPSILON
    Done=0                                   ## SET DONE TO ZERO (BEGINNING OF THE EPISODE)
    i = 1                                    ## COUNTER FOR NUMBER OF ACTIONS
    
    S,X,allowed_a=env.Initialise_game()      ## INITIALISE GAME
                                     
    while Done==0: 
        ## START THE EPISODE
        
        # Neural activation: input layer -> hidden layer
        h1 = np.dot(W1,X)+bias_W1

        # Apply the ReLu function
        x1 = relu(h1)

        # Neural activation: hidden layer -> output layer
        h2 = np.dot(W2,x1)+bias_W2

        # Apply the ReLu function
        x2 = relu(h2)
            
        #choose action :
        # Epsilon-greedy parameter
        # with probability epsilon choose action at random; if epsilon = 0 then always choose Greedy
        eGreedy = int(np.random.rand() < epsilon_f) 

        # Implement the policy
        if  eGreedy:
            # if  epsilon>0 (e-Greedy) choose one at random
            a,_=np.where(allowed_a==1) 
            a_agent=np.random.permutation(a)[0]
        else:
            # otherwise choose greedy
            # will result the best Qvalue allowed action
            a,_=np.where(allowed_a==1) 
            Q_allowed=np.empty(0)
            for h in a:
                Q_allowed=np.append(Q_allowed,x2[h])   ##Qvalue only for allowed actions
            Q_a_agent=np.argmax(Q_allowed) 
            a_agent=a[Q_a_agent]                       ##action we choose
        
        # Collect reward
        Sprime,Xprime,allowed_aprime,R,Done=env.OneStep(a_agent)
        
        # Create a mask that only updates the action taken
        mask = np.zeros_like(N_out)
        mask[0, a_agent] = 1


        
      
        
        ## THE EPISODE HAS ENDED, UPDATE
        
        if Done==1:
            # Compute the error signal
            e_n = R-Q_a_agent
                
            # Backpropagation: output layer -> hidden layer
            delta2 = np.heaviside(h2, 0) * e_n * mask
                
            dW2 = np.outer(delta2, x1)
            dbias_W2 = delta2

            # Backpropagation: hidden layer -> input layer
            delta1 = np.heaviside(h1,0.5) * np.dot(W2.T, delta2)
            dW1 = np.outer(delta1,X)
            dbias_W1 = delta1
            
            #  update the weights using accumulated gradients
            W2 += eta*dW2
            W1 += eta*dW1

            bias_W1 += eta*dbias_W1
            bias_W2 += eta*dbias_W2 
            
            
            #record reward and number of step
            R_save[n] = R
            N_moves_save[n] = i
            break
        
        
        
        
        # IF THE EPISODE IS NOT OVER...
        else:
            #choose a' from s' using policy derived from Q
        
            # Neural activation: input layer -> hidden layer
            h1prime = np.dot(W1,Xprime)+bias_W1

            # Apply the ReLu function
            x1prime = relu(h1prime)

            # Neural activation: hidden layer -> output layer
            h2prime = np.dot(W2,x1prime)+bias_W2

            # Apply the ReLu function
            x2prime = relu(h2prime)
            
            #choose action :
            # Epsilon-greedy parameter
            # with probability epsilon choose action at random; if epsilon = 0 then always choose Greedy
            eGreedy = int(np.random.rand() < epsilon_f) 
        
            # Implement the policy
            if  eGreedy:
                # if  epsilon>0 (e-Greedy) choose one at random
                a,_=np.where(allowed_aprime==1) 
                a_agentprime=np.random.permutation(a)[0]
                Q_a_agentprime=x2[a_agentprime]
            else:
                # otherwise choose greedy
                # will result the best Qvalue allowed action
                a,_=np.where(allowed_aprime==1) 
                Q_allowedprime=np.empty(0)
                for h in a:
                    Q_allowedprime=np.append(Q_allowedprime,x2prime[h])      ##Qvalue only for allowed actions
                Q_a_agentprime=np.argmax(Q_allowedprime)
                a_agentprime=a[Q_a_agentprime]                               ##action we choose
                
                
                
            
            # Compute the error signal
            e_n = R+gamma*Q_a_agentprime - Q_a_agent
                
            # Backpropagation: output layer -> hidden layer
            delta2 = np.heaviside(h2, 0) * e_n * mask
                
            dW2 = np.outer(delta2, x1)
            dbias_W2 = delta2

            # Backpropagation: hidden layer -> input layer
            delta1 = np.heaviside(h1,0.5) * np.dot(W2.T, delta2)
            dW1 = np.outer(delta1,X)
            dbias_W1 = delta1
            
            #  update the weights using accumulated gradients
            W2 += eta*dW2
            W1 += eta*dW1

            bias_W1 += eta*dbias_W1
            bias_W2 += eta*dbias_W2           
        

            
            
        # NEXT STATE AND CO. BECOME ACTUAL STATE...     
        S=np.copy(Sprime)
        X=np.copy(Xprime)
        allowed_a=np.copy(allowed_aprime)
        
        i += 1  # UPDATE COUNTER FOR NUMBER OF ACTIONS

        
        
    
    
```

```python
#training with Q-learning Algorithm
#we do this for every game
for n in range(N_episodes):

    epsilon_f = epsilon_0 / (1 + beta * n)   ## DECAYING EPSILON
    Done=0                                   ## SET DONE TO ZERO (BEGINNING OF THE EPISODE)
    i = 1                                    ## COUNTER FOR NUMBER OF ACTIONS
    
    S,X,allowed_a=env.Initialise_game()      ## INITIALISE GAME
                                     
    while Done==0: 
        ## START THE EPISODE
        
        
        # Neural activation: input layer -> hidden layer
        h1 = np.dot(W1,X)+bias_W1

        # Apply the ReLu function
        x1 = relu(h1)

        # Neural activation: hidden layer -> output layer
        h2 = np.dot(W2,x1)+bias_W2

        # Apply the ReLu function
        x2 = relu(h2)
            
        #choose action :
        # Epsilon-greedy parameter
        # with probability epsilon choose action at random; if epsilon = 0 then always choose Greedy
        eGreedy = int(np.random.rand() < epsilon_f) 

        # Implement the policy
        if  eGreedy:
            # if  epsilon>0 (e-Greedy) choose one at random
            a,_=np.where(allowed_a==1) 
            a_agent=np.random.permutation(a)[0]
        else:
            # otherwise choose greedy
            # will result the best Qvalue allowed action
            a,_=np.where(allowed_a==1) 
            Q_allowed=np.empty(0)
            for h in a:
                Q_allowed=np.append(Q_allowed,x2[h])        ##Qvalue for only allowed actions
            Q_a_agent=np.argmax(Q_allowed) 
            a_agent=a[Q_a_agent]                            ##action we choose
        
        # Collect reward
        Sprime,Xprime,allowed_aprime,R,Done=env.OneStep(a_agent)
        
        # Create a mask that only updates the action taken
        mask = np.zeros_like(N_out)
        mask[0, a_agent] = 1


        
      
        
        ## THE EPISODE HAS ENDED, UPDATE...BE CAREFUL, THIS IS THE LAST STEP OF THE EPISODE
        
        if Done==1:
            # Compute the error signal
            e_n = R-Q_a_agent
                
            # Backpropagation: output layer -> hidden layer
            delta2 = np.heaviside(h2, 0) * e_n * mask
                
            dW2 = np.outer(delta2, x1)
            dbias_W2 = delta2

            # Backpropagation: hidden layer -> input layer
            delta1 = np.heaviside(h1,0.5) * np.dot(W2.T, delta2)
            dW1 = np.outer(delta1,X)
            dbias_W1 = delta1
            
            #  update the weights using accumulated gradients
            W2 += eta*dW2
            W1 += eta*dW1

            bias_W1 += eta*dbias_W1
            bias_W2 += eta*dbias_W2 
            
            
            #record reward and number of step
            R_save[n] = R
            N_moves_save[n] = i
            break
        
        
        
        
        # IF THE EPISODE IS NOT OVER...
        else:
            #find max(Q)
            #for this we need to calculate all Q
        
            # Neural activation: input layer -> hidden layer
            h1prime = np.dot(W1,Xprime)+bias_W1

            # Apply the ReLu function
            x1prime = relu(h1prime)

            # Neural activation: hidden layer -> output layer
            h2prime = np.dot(W2,x1prime)+bias_W2

            # Apply the ReLu function
            x2prime = relu(h2prime)
            
            
            #find Qmax :
               
            # will result the best Qvalue 
            a,_=np.where(allowed_aprime==1) 
            Q_allowedprime=np.empty(0)
            for h in a:
                Q_allowedprime=np.append(Q_allowedprime,x2prime[h])       ## Qvalue for alllowed action only
            Q_max=np.argmax(Q_allowedprime)                               ##that's max(Q) for next state in allowed action
                
                
                
            
            # Compute the error signal
            e_n = R+gamma*Q_max - Q_a_agent
                
            # Backpropagation: output layer -> hidden layer
            delta2 = np.heaviside(h2, 0) * e_n * mask
                
            dW2 = np.outer(delta2, x1)
            dbias_W2 = delta2

            # Backpropagation: hidden layer -> input layer
            delta1 = np.heaviside(h1,0.5) * np.dot(W2.T, delta2)
            dW1 = np.outer(delta1,X)
            dbias_W1 = delta1
            
            #  update the weights using accumulated gradients
            W2 += eta*dW2
            W1 += eta*dW1

            bias_W1 += eta*dbias_W1
            bias_W2 += eta*dbias_W2           
        

            
            
        # NEXT STATE AND CO. BECOME ACTUAL STATE...     
        S=np.copy(Sprime)
        X=np.copy(Xprime)
        allowed_a=np.copy(allowed_aprime)
        
        i += 1  # UPDATE COUNTER FOR NUMBER OF ACTIONS
```

```python


print(R_save)
# Plot the average reward as a function of the number of trials 
# --> the average has to be performed over the episodes
reward=EMA(R_save,1000)
plt.figure(figsize = (8, 6))
plt.plot(np.arange(0,N_episodes,1),reward)

plt.xlabel('Trial')
plt.ylabel('Average Reward')
plt.axis((-(N_episodes/10.0),N_episodes,-0.1,1.1))
plt.tick_params(axis = 'both', which='major', labelsize = 14)
plt.show()
```
